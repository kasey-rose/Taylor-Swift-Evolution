## Cleaning and Preparing Data

importing the packages that we're going to need:

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

importing the data for each album as a data frames:

```python
debut = pd.read_csv("01-taylor_swift.csv")
speaknow = pd.read_csv("03-speak_now_deluxe_package.csv")
red = pd.read_csv("04-red_deluxe_edition.csv")
nineteen89 = pd.read_csv("05-1989_deluxe.csv")
rep = pd.read_csv("06-reputation.csv")
lover = pd.read_csv("07-lover.csv")
folk = pd.read_csv("08-folklore_deluxe_version.csv")
ever = pd.read_csv("09-evermore_deluxe_version.csv")
fearless = pd.read_csv("02-fearless_taylors_version.csv")
```

renaming columns in the data frames:

```python
debut.rename({"album_name" : "Album", "track_title":"Track Title", "track_n":"Track Number"}, 
             axis = 1, inplace = True)

speaknow.rename({"album_name" : "Album", "track_title":"Track Title", "track_n":"Track Number"}, 
                axis = 1, inplace = True)

red.rename({"album_name" : "Album", "track_title":"Track Title", "track_n":"Track Number"}, 
           axis = 1, inplace = True)

nineteen89.rename({"album_name" : "Album", "track_title":"Track Title", "track_n":"Track Number"}, 
                  axis = 1, inplace = True)

rep.rename({"album_name" : "Album", "track_title":"Track Title", "track_n":"Track Number"}, 
           axis = 1, inplace = True)

lover.rename({"album_name" : "Album", "track_title":"Track Title", "track_n":"Track Number"}, 
             axis = 1, inplace = True)

folk.rename({"album_name" : "Album", "track_title":"Track Title", "track_n":"Track Number"}, 
            axis = 1, inplace = True)

ever.rename({"album_name" : "Album", "track_title":"Track Title", "track_n":"Track Number"}, 
            axis = 1, inplace = True)

fearless.rename({"album_name" : "Album", "track_title":"Track Title", "track_n":"Track Number"}, 
                axis = 1, inplace = True)
```

## How has the number of lines in Taylor Swift's songs changed over time?

### Looking at Each Song Individually:

setting a variable for each album's columns "Track Title"

counting how frequently each track title appears, and sets that to the number of lines

creating a data frame with two columns, the titles of the tracks, and the number of lines in the track

```python
debut_tracks = debut["Track Title"]
debut_lines_per_song = debut_tracks.value_counts()
debut_songs = pd.DataFrame(debut_lines_per_song.reset_index().values, columns = ["Track Name", "Number of Lines"])

speaknow_tracks = speaknow["Track Title"]
speaknow_lines_per_song = speaknow_tracks.value_counts()
speaknow_songs = pd.DataFrame(speaknow_lines_per_song.reset_index().values, columns = ["Track Name", "Number of Lines"])

red_tracks = red["Track Title"]
red_lines_per_song = red_tracks.value_counts()
red_songs = pd.DataFrame(red_lines_per_song.reset_index().values, columns = ["Track Name", "Number of Lines"])

nineteen89_tracks = nineteen89["Track Title"]
nineteen89_lines_per_song = nineteen89_tracks.value_counts()
nineteen89_songs = pd.DataFrame(nineteen89_lines_per_song.reset_index().values, columns = ["Track Name", "Number of Lines"])

rep_tracks = rep["Track Title"]
rep_lines_per_song = rep_tracks.value_counts()
rep_songs = pd.DataFrame(rep_lines_per_song.reset_index().values, columns = ["Track Name", "Number of Lines"])

lover_tracks = lover["Track Title"]
lover_lines_per_song = lover_tracks.value_counts()
lover_songs = pd.DataFrame(lover_lines_per_song.reset_index().values, columns = ["Track Name", "Number of Lines"])

folk_tracks = folk["Track Title"]
folk_lines_per_song = folk_tracks.value_counts()
folk_songs = pd.DataFrame(folk_lines_per_song.reset_index().values, columns = ["Track Name", "Number of Lines"])

ever_tracks = ever["Track Title"]
ever_lines_per_song = ever_tracks.value_counts()
ever_songs = pd.DataFrame(ever_lines_per_song.reset_index().values, columns = ["Track Name", "Number of Lines"])

fearless_tracks = fearless["Track Title"]
fearless_lines_per_song = fearless_tracks.value_counts()
fearless_songs = pd.DataFrame(fearless_lines_per_song.reset_index().values, columns = ["Track Name", "Number of Lines"])
```

creating a list containing all songs from the dataframes of the songs in each album

creating a data frame of lines per track and resetting the index 

```python
songs = [debut_songs, speaknow_songs, red_songs, nineteen89_songs, rep_songs, lover_songs, folk_songs, ever_songs, fearless_songs]
lines_per_track = pd.concat(songs)
lines_per_track.reset_index(inplace=True)
```

creating a list of colors, and for the number of songs in each album, there is the same amount of the corresponding album colors

```python
colors = []

for i in debut_songs["Track Name"]:
    colors.append("mediumturquoise")

for i in speaknow_songs["Track Name"]:
    colors.append("rebeccapurple")
        
for i in red_songs["Track Name"]:
    colors.append("firebrick")

for i in nineteen89_songs["Track Name"]:
    colors.append("dodgerblue")
        
for i in rep_songs["Track Name"]:
    colors.append("darkslategrey")
        
for i in lover_songs["Track Name"]:
    colors.append("deeppink")
        
for i in folk_songs["Track Name"]:
    colors.append("lightslategray")     

for i in ever_songs["Track Name"]:
     colors.append("darkolivegreen")
        
for i in fearless_songs["Track Name"]:
     colors.append("darkgoldenrod")
```

creating a dictionary of the album name and it's corresponding color

creating a list with the years of releaee of each album

```python
album_colors = {"Taylor Swift (2006)" : "mediumturquoise", "Speak Now (2010)" : "rebeccapurple",
                "Red (2012)" : "firebrick", "1989 (2014)" : "dodgerblue",
                "Reputation (2017)" : "darkslategrey", "Lover (2019)" : "deeppink",
                "Folklore (2020)" : "lightslategray", "Evermore (2020)" : "darkolivegreen",
                "Fearless (Taylor's Version) (2021)" : "darkgoldenrod"}

year_of_release = [2008, 2010, 2012, 2014, 2017, 2019, 2020, 2020, 2021]
```

creating a bar graph:

```python
fig, ax = plt.subplots()
names = lines_per_track["Track Name"]
number = lines_per_track["Number of Lines"]
ax.bar(names, number, color = colors, width = .5)

ax.xaxis.set_tick_params(labelbottom=False)
ax.set_xlabel("Taylor Swift Songs (in Chronological Order of Release)")
ax.set_ylabel("Number of Lines in the Song")
ax.set_title("Number of Lines In Taylor Swift Songs")

labels = list(album_colors.keys())
handles = [plt.Rectangle((0,0),1,1, color=album_colors[label]) for label in labels]
plt.legend(handles, labels, bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)

plt.show()
```

#### The Average Number of Lines per Song in the Album

creating variables representing the average number of lines per song in each song

putting the averages into a single variable as a list

creating a two lists of strings: one is the names of the albums and one is the colors that correspond to each album

```python
avg_debut_lines_per_song = debut_lines_per_song.mean()
avg_speaknow_lines_per_song = speaknow_lines_per_song.mean()
avg_red_lines_per_song = red_lines_per_song.mean()
avg_nineteen89_lines_per_song = nineteen89_lines_per_song.mean()
avg_rep_lines_per_song = rep_lines_per_song.mean()
avg_lover_lines_per_song = lover_lines_per_song.mean()
avg_folk_lines_per_song = folk_lines_per_song.mean()
avg_ever_lines_per_song = ever_lines_per_song.mean()
avg_fearless_lines_per_song = fearless_lines_per_song.mean()

avg_lines_per_track = [avg_debut_lines_per_song, avg_speaknow_lines_per_song, avg_red_lines_per_song,
                       avg_nineteen89_lines_per_song, avg_rep_lines_per_song, avg_lover_lines_per_song,
                       avg_folk_lines_per_song, avg_ever_lines_per_song, avg_fearless_lines_per_song]

albums = ["Taylor Swift (2006)", "Speak Now (2010)", "Red (2012)", "1989 (2014)", "Reputation (2017)",
          "Lover (2019)", "Folklore (2020)", "Evermore (2020)", "Fearless (Taylor's Version) (2021)"]

avg_colors = ["mediumturquoise", "rebeccapurple", "firebrick", "dodgerblue", "darkslategrey", "deeppink", 
              "lightslategray", "darkolivegreen", "darkgoldenrod"]
```

creating a bar graph with the average number of lines

```python
fig, ax = plt.subplots()
ax.bar(albums, avg_lines_per_track, color = avg_colors, width = .5)

ax.xaxis.set_tick_params(labelbottom=False)
ax.set_xlabel("Taylor Swift Albums (in Chronological Order of Release)")
ax.set_ylabel("Average Number of Lines of the Songs in the Album")
ax.set_title("Average Number of Lines In Taylor Swift Songs")

labels = list(album_colors.keys())
handles = [plt.Rectangle((0,0),1,1, color=album_colors[label]) for label in labels]
plt.legend(handles, labels, bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)

plt.show()
```

creating a bar graph with the average number of lines

```python
xy = list(zip(year_of_release, avg_lines_per_track))
xy = np.array(xy)
x = xy[:, 0]
y = xy[:, 1]



theta = np.polyfit(x, y, 1)
r = theta[1] + theta[0]*x
plt.scatter(x, y, color = "midnightblue")
plt.plot(x, r, color = "midnightblue")

plt.title("Percentage of Songs by Taylor Swift Per Album, by Year")
plt.xlabel('Year of Album Release')
plt.ylabel("Percent of Songs in the Album out of Swift's entire discography")

plt.show()
```

changing the list of songs in the album that follow the category into a set to remove duplicates

creating a variable for each album, representing the percent of songs in the album in the category out of all of the songs on the album

```python
debut_love_songs = [*set(debut_love_songs)]
speaknow_love_songs = [*set(speaknow_love_songs)]
red_love_songs = [*set(red_love_songs)]
nineteen89_love_songs = [*set(nineteen89_love_songs)]
rep_love_songs = [*set(rep_love_songs)]
lover_love_songs = [*set(lover_love_songs)]
folk_love_songs = [*set(folk_love_songs)]
ever_love_songs = [*set(ever_love_songs)]
fearless_love_songs = [*set(fearless_love_songs)]

debut_percent_love = len(debut_love_songs)/len(debut_tracklist)
speaknow_percent_love = len(speaknow_love_songs)/len(speaknow_tracklist)
red_percent_love = len(red_love_songs)/len(red_tracklist)
nineteen89_percent_love = len(nineteen89_love_songs)/len(nineteen89_tracklist)
rep_percent_love = len(rep_love_songs)/len(rep_tracklist)
lover_percent_love = len(lover_love_songs)/len(lover_tracklist)
folk_percent_love = len(folk_love_songs)/len(folk_tracklist)
ever_percent_love = len(ever_love_songs)/len(ever_tracklist)
fearless_percent_love = len(fearless_love_songs)/len(fearless_tracklist)
```
